```python
import sympy as sym
from sympy import symbols, latex, diff, S
from sympy.solvers import solve
from sympy.plotting import plot

sym.init_printing(use_latex='mathjax')
```

```python
x = symbols('x',real=True) # ,real=True
f = x**2 - 6*abs(x) + 8
f
```

```python
print('корни функции')
solve(f,x)
```

```python
f1 = diff(f)
f1 # производная
```

```python
extremum_x = solve(f1,x) # значение X в точке экстремума
extremum_x[0].n()
```

```python
print('Функция симметричная (в ней есть модуль) - то точка экстремума так же будет симметрична')
print(f.subs(x,-extremum_x[0]).n()) #
print(f.subs(x,0).n())
print(f.subs(x,extremum_x[0]).n())
```

```python
print(f1.subs(x,-2), 'На промежутке от -oo до -extremum_x[0] (-oo;-3) - производная отрицательная - функция f(x) - убывает')
print(f1.subs(x,-0.2), 'На промежутке от -extremum_x[0] до 0 (-3;0) - производная положительная - функция f(x) - возрастает')
print(f1.subs(x,0.2), 'На промежутке от 0 до extremum_x[0] (0;3) - производная отрицательная - функция f(x) - убывает')
print(f1.subs(x,2), 'На промежутке от extremum_x[0] до +oo (3;+oo) - производная положительная - функция f(x) - возрастает')
```

```python
print('строим график')
str= latex(S(f,evaluate = False))
p = plot(f,(x,-8,8),title='$'+'f(x)='+str+'$', show=  False)
p.show()
```

```python
print('находим вершину')
str= latex(S(f1,evaluate = False))
p = plot(f1,(x,-5,5), title = '$'+'f1(x)='+str+'$', show = False) # график производной
p.show()
```
